2. Buat sebuah program OOP dengan ketentuan:

  - berbeda dari contoh program yang ada di materi (ppt dan colab)
  - memiliki 1 parent class
  - memiliki setidaknya 2 child class
  - menerapkan konsep dasar OOP (kecuali abstraction)
  - dilengkapi dengan docstring, contoh docstring bisa dilihat dalam program hero.py

  Setelah program selesai, buatlah:
  - setidaknya 2 instance untuk setiap child class
  - Class Diagram-nya

```python
# parent class 
class provider:
  """ 
  Nama
    provider
  Deskripsi
    Mengetahui data diri pada provider GSM, yaitu Telkomsel, dan XL.
  -Dapat menentukan banyaknya pelanggan
  -Dapat menentukan Pengguna kartu apa
  """

  provider = 0
  
  def pelanggan(self):  
    print('menentukan banyaknya pelanggan yang digunakan')

  def get_pelanggan(self):
    print(f"""{self.nama}, kamu adalah salah satu pengguna user {self.operator} jenis provider yang banyak digunakan pada kartu {self.jenis}""")
    print('Memiliki pengguna sebanyak :')
    return self.pelanggan


# child class 1 (byu)
class Telkomsel(provider): 
  def __init__(self,nama, operator, jenis, total):

    self.nama = nama
    self.operator = operator
    self.jenis = jenis
    self.total = total

  def pelanggan(self):  
    self.pelanggan = self.total="162.048Juta"

# child class 1 (Indosat)
class indosat(provider): 
  def __init__(self,nama, operator, jenis, total):

    self.nama = nama
    self.operator = operator
    self.jenis = jenis
    self.total = total

  def pelanggan(self):  
    self.pelanggan = self.total = "55.054 Juta"
```

```python
Indosat1 =indosat('Fajri','Indosat Ooredo','prabayar',"total")
Indosat1.pelanggan()
Indosat1.get_pelanggan()
```

```python
telkomsel1 = Telkomsel('Fadil','Tsel', 'prabayar',"total")
telkomsel1.pelanggan()
telkomsel1.get_pelanggan()
```
